##### **PROBLEM 1** - Create the following data frames based on the format provided:
##### b. Filename: Mindy = [ “Name”, “Track”, “Electronics”, “Average >=55”]; where hometown is constant as Mindanao and gender Female
##### *Expected Outcome: Data Wrangling of Name, Track, Elctronics, and an Average greater than or equal to 55 with Hometown-Mindanao and Gender-Female as constant*

```python
#Pandas function to average specific columns in a row
boards['Ave']=boards[['Math','Electronics','GEAS','Communication']].mean(axis=1)
#Pandas function to create a dataframe from the loaded xlsx file
Mindy = boards.loc[(boards['Hometown']=='Mindanao')
                &(boards['Gender']=='Female'),
                ['Name','Track','Electronics','Ave']]
Mindy_filtered = Mindy[Mindy['Ave'] >= 55] #Filters the column of Averahe that has a value greater than or equal to 55
Mindy_filtered
```

##### **PROBLEM 2** - Create a visualization that shows how the different features contributes to average grade. Does chosen track in college, gender, or hometown contributes to a higher average score? 
##### *Expected Outcome: A graph of how different features contribute to the average grade*

```python
combined_avg = boards.groupby(['Track', 'Gender', 'Hometown'])['Ave'].mean().unstack()
print("\nCombined Average Grade by Track, Gender, and Hometown:")
print(combined_avg)

fig, axes = plt.subplots(3, 1, figsize=(10, 18))

# Average Grade by Track
boards.groupby('Track')['Ave'].mean().plot(kind='bar', ax=axes[0], color='#A6B37D', title='Average Grade by Track', ylabel='Average Grade')

# Average Grade by Gender
boards.groupby('Gender')['Ave'].mean().plot(kind='bar', ax=axes[1], color='#B99470', title='Average Grade by Gender', ylabel='Average Grade')

# Average Grade by Hometown
boards.groupby('Hometown')['Ave'].mean().plot(kind='bar', ax=axes[2], color='#FF7777', title='Average Grade by Hometown', ylabel='Average Grade')

# Adjust layout
plt.tight_layout()

# Show the combined plot
plt.show()
```
